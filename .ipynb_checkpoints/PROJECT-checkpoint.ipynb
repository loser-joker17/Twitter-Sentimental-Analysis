```python
#np.loadtxt("0000000000002747_training_twitter_x_y_train.csv")
#text=.raw("0000000000002747_training_twitter_x_y_train.csv")
data=pd.read_csv("0000000000002747_training_twitter_x_y_train.csv")
data_test=pd.read_csv("0000000000002747_test_twitter_x_test.csv")
```

```python
b=data["airline_sentiment"]
#b_test=data_test["airline_sentiment"]
```

```python
a=data["text"]
a_test=data_test["text"]
a, b,a_test
```

```python
from nltk.tokenize import word_tokenize
```

```python
document=[]
for x in a:
    document.append(word_tokenize(x))
doc_test=[]
for x in a_test:
    doc_test.append(word_tokenize(x))
```

```python
document[0], doc_test[0]
```

```python
categories=[]
for x in b:
    categories.append(word_tokenize(x)[0])
```

```python
categories
```

```python
documents=[]
i=0
while i<len(categories):
    documents.append((document[i],categories[i]))
    i=i+1
```

```python
documents[0]
```

```python
from nltk.corpus import stopwords
stops=set(stopwords.words("english"))
import string
punctuations=list(string.punctuation)
stops.update(punctuations)
stops
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
from nltk import pos_tag
from nltk.corpus import wordnet
```

```python
def get_simple_pos(tag):
    if tag.startswith("J"):
        return wordnet.ADJ
    elif tag.startswith("V"):
        return wordnet.VERB
    elif tag.startswith("N"):
        return wordnet.NOUN
    elif tag.startswith("R"):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def clean_data(words):
    output_words=[]
    for w in words:
        if w.lower() not in stops:
            pos=pos_tag(w)
            clean_word=lemmatizer.lemmatize(w, pos=get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents=[(clean_data(document),category) for document, category in documents]
```

```python
doc_test=[clean_data(document) for document in doc_test]
```

```python
documents[0], doc_test[0]
```

# Building Feature Set

```python
all_words=[]
for doc in documents:
    all_words+=doc[0]
```

```python
all_words
```

```python
import nltk
Freq = nltk.FreqDist(all_words)
Freq
```

```python
common=Freq.most_common(15)
common
```

```python
features=[ i[0] for i in common]
features
```

```python
def get_feature_dict(words):
    current_features={}
    words_set=set(words)
    for w in features:
        current_features[w]=w in words_set
    return current_features
```

```python
training_data=[(get_feature_dict(doc),category) for doc, category in documents]
```

```python
training_data[0]
```

# Classification Using NLTK Naive Bayes

```python
from nltk import NaiveBayesClassifier
classifier=NaiveBayesClassifier.train(training_data)
```

```python
#y_pred=classifier.predict(doc_test)
```

```python
#classifier.classify(doc_test)
```

```python
classifier.show_most_informative_features(15)
```

```python
training_data[0], doc_test[0]
```

```python
import nltk
w=[]
for x in doc_test:
    Freq = nltk.FreqDist(x)
    #t=get_feature_dict(Freq)
    w.append(classifier.classify(Freq))
```

```python
w
```

```python
testing_data_a=[]
i=0
while i<len(w):
    testing_data_a.append((doc_test[i],w[i]))
    i=i+1
```

```python
testing_data_a[0]
#,len(doc_test), len(w)
testing_data=[(get_feature_dict(doc),category) for doc, category in testing_data_a]
```

```python
training_data[0],testing_data[0]
```

```python
nltk.classify.accuracy(classifier,testing_data)
```
